```python
import pandas as pd
import numpy as np
```

```python
rng = np.random.default_rng(42)
```

```python
runs = pd.read_csv('data/ENA_runs.annotated.csv.gz')
```

```python
samples = runs.groupby('sample_accession').first()
samples['sample_read_count'] = runs.groupby('sample_accession').sum()['read_count']
```

# Transcriptomic

```python
taxids = [6239, 10090, 7227, 4932, 9606]
```

```python
txomes = samples[samples.library_source == 'TRANSCRIPTOMIC']
```

```python
sampled_txomes = []
for tid in taxids:
    print(tid)
    sampled_txomes.append(txomes[txomes.tax_id == tid].sample(n=3, random_state=rng))
sampled_txomes = pd.concat(sampled_txomes)
```

# Genomics

```python
genomes = samples[samples.library_source == 'GENOMIC']
```

```python
sampled_genomes = []
for tid in taxids:
    print(tid)
    sampled_genomes.append(genomes[genomes.tax_id == tid].sample(n=3, random_state=rng))
sampled_genomes = pd.concat(sampled_genomes)
```
